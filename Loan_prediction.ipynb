```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset\\train_u6lujuX_CVtuZ9i.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# Checking if there are any missing values in any of the features
df.isnull().sum()
```

```python
df
```

```python
cols = ['Gender','Married','Dependents','Self_Employed','Loan_Amount_Term','Credit_History']

for c in cols:
    df[c].fillna(df[c].mode()[0], inplace=True)

df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)
```

```python
# Filled all the missing values
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cat_cols = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area']

for c in cat_cols:
    df[c] = le.fit_transform(df[c])

df['Loan_Status'] = le.fit_transform(df['Loan_Status'])
```

```python
# Independent Features
X = df[['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']]

# Dependent Feature
y = df['Loan_Status']
```

```python
X.head()
```

```python
y.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train
```

```python
y_test
```

```python
# Standardization 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
nums_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']

X_train[nums_cols] = scaler.fit_transform(X_train[nums_cols])
X_test[nums_cols] = scaler.transform(X_test[nums_cols])
```

```python
X_train
```

```python
# Training different models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_lr = lr.predict(X_test)
y_pred_knn = knn.predict(X_test)
y_pred_dt = dt.predict(X_test)
```

```python
# Performance Metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Logistic Regression Performance
print(accuracy_score(y_test, y_pred_lr))
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
# KNN Performance
print(accuracy_score(y_test, y_pred_knn))
print(confusion_matrix(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

```python
# Decision Tree Performance
print(accuracy_score(y_test, y_pred_dt))
print(confusion_matrix(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
```

```python
# GridSearchCV Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'C': [0.01, 0.1, 1, 5, 10, 50, 100],
    'solver': ['liblinear', 'lbfgs', 'newton-cg', 'saga'],
    'penalty': ['l1', 'l2', 'elasticnet'],
    'max_iter': [200, 500, 800]    
}
```

```python
grid = GridSearchCV(estimator=lr, param_grid=params, scoring='accuracy', cv=5, n_jobs=-1)
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
print(grid.best_score_)
```

```python
# GridSearchCV Performance Metrics
y_pred_grid = grid.predict(X_test)
print(accuracy_score(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
print(classification_report(y_test, y_pred_grid))
```
